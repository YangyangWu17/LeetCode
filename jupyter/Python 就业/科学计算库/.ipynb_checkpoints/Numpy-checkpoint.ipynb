# 画图

```python
x=np.random.uniform(0,1,1000000)

plt.figure(figsize=(20,8),dpi=100)

plt.hist(x,bins=100)

plt.show()
```

```python
x2=np.random.normal(0,1,1000000)

plt.figure(figsize=(20,8),dpi=100)

plt.hist(x2,bins=1000)

plt.show()
```

```python
stock_change = np.random.normal(0, 1, (4,5))
stock_change
```
