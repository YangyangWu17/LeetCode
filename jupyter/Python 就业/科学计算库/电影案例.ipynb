```python
plt.figure(figsize=(20,8),dpi=100)
plt.hist(movie['Rating'].values,20)

max=movie['Rating'].max()
min=movie['Rating'].min()

x1=np.linspace(min,max,21)

plt.xticks(x1)

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(20,8),dpi=100)
plt.hist(movie['Runtime (Minutes)'].values,20)

max=movie['Runtime (Minutes)'].max()
min=movie['Runtime (Minutes)'].min()

x1=np.linspace(min,max,21)

plt.xticks(x1)

plt.grid()

plt.show()
```

```python
temp_list=[i.split(',')for i in movie['Genre']]
```

```python
list=np.unique([j for i in temp_list for j in i])
```

```python
list.shape
```

```python
genre=pd.DataFrame(np.zeros([movie.shape[0],list.shape[0]]),columns=list)
```

```python
genre.head()
```

```python
for i in range(1000):
    genre.loc[i,temp_list[i]]=1
```

```python
genre
```

```python
genre.sum().sort_values(ascending=False).plot(kind="bar",colormap='cool',figsize=(20,8),fontsize=18)
```
