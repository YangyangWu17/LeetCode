# 折线图


```python
x=range(60)
y_bj=[random.uniform(10,15) for i in x]
y_sh=[random.uniform(20,25) for i in x]

#创建画布
plt.figure(figsize=(20,8),dpi=100)

#绘制图像
plt.plot(x,y_bj,label='bj')
plt.plot(x,y_sh,label='sh',color='g',linestyle='-.')
)

#2.1添加x,y轴刻度
x_ticks_label=["10.{}s".format(i)for i in x]
y_ticks=range(40)

plt.xticks(x[::5],x_ticks_label[::5])
plt.yticks(y_ticks[::5])

#添加网格
plt.grid(True,linestyle='-',alpha=0.2)

#添加描述
plt.xlabel("t")
plt.ylabel("temp")
plt.title("one hour",fontsize=20)

#显示图例
plt.legend(loc=3)

#显示图像  
plt.show()
```

# 常见图形绘制


```python
#柱状图bar
#直方图hist
#饼图pie
#散点图  scatter

#数据准备
x=[1,2,3,4,5]
y=[2,44,5,6,223]
#创建画布
plt.figure(figsize=(20,8),dpi=100)
#图像绘制
plt.scatter(x,y)
#图像展示
plt.show()
```

```python
#柱状图
movie_name=['lalalal','dadada','aaa']

x=range(len(movie_name))
y=[113421,2242,55664]

plt.figure(figsize=(20,8),dpi=100)

plt.bar(x,y,width=0.5,color=['b','y','r'])

plt.xticks(x,movie_name)

plt.grid(linestyle='--',alpha=0.5)

plt.title('sss')

plt.show()
```
